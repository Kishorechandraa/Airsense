```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('AQI.csv')
df
```

```python
from sklearn.model_selection import train_test_split

x=df.iloc[:,2:14].values
y=df.iloc[:,-2].values

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
print(f'X_train shape: {X_train.shape}')
print(f'X_test shape: {X_test.shape}')
print(f'y_train shape: {y_train.shape}')
print(f'y_test shape: {y_test.shape}')
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score


model_linear = LinearRegression()
model_linear.fit(X_train, y_train)

y_pred_linear = model_linear.predict(X_test)

mse_linear = mean_squared_error(y_test, y_pred_linear)
mae_linear = mean_absolute_error(y_test, y_pred_linear)
r2_linear = r2_score(y_test, y_pred_linear)
rmse_linear = mse_linear ** 0.5

print(f"Linear Regression MSE: {mse_linear}")
print(f"Linear Regression MAE: {mae_linear}")
print(f"Linear Regression RMSE: {rmse_linear}")
print(f"Linear Regression RÂ² Score: {r2_linear}")
```

Linear Regression Metrics:

Mean Squared Error (MSE):
Value: 1.735979448002134
Interpretation: This is a very low MSE, indicating that the predictions are very close to the actual values on average. The model's error is quite small.

Mean Absolute Error (MAE):
Value: 0.0031087309638913223
Interpretation: This low MAE suggests that the average error of the predictions is approximately 0.0031 units. The model is making very accurate predictions.

Root Mean Squared Error (RMSE):
Value: 0.004166508667940262
Interpretation: The RMSE is also very low, indicating that the average distance between the predicted and actual values is minimal.

RÂ² Score:
Value: 0.9999999983400809
Interpretation: This value is very close to 1, suggesting that the linear regression model explains nearly all the variability in the data. It indicates an excellent fit.

```python
from sklearn.neighbors import KNeighborsRegressor

model_knn = KNeighborsRegressor(n_neighbors=7)
model_knn.fit(X_train, y_train)

y_pred_knn = model_knn.predict(X_test)

mse_knn = mean_squared_error(y_test, y_pred_knn)
mae_knn = mean_absolute_error(y_test, y_pred_knn)
r2_knn = r2_score(y_test, y_pred_knn)
rmse_knn = mse_knn ** 0.5

print(f"KNN Regression MSE: {mse_knn}")
print(f"KNN Regression MAE: {mae_knn}")
print(f"KNN Regression RMSE: {rmse_knn}")
print(f"KNN Regression RÂ² Score: {r2_knn}")
```
